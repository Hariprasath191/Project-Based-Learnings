```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head(5)
```

```python
plt.scatter(data.StudyTime,data.Score)
plt.show()
```

```python
def loss_function(m,b,points):
    total_error = 0 
    for  i in range(len(points)):
        x = data.iloc[i].StudyTime
        y = data.iloc[i].Score 
        total_error += ( y -( m*x + b ))**2
    return total_error/float(len(points))
```

```python
def gradient_descent(m_now,b_now,points,L):
    m_gradient = 0
    b_gradient = 0 
    n = len(points)

    for i in range(n):
        x = data.iloc[i].StudyTime
        y = data.iloc[i].Score 

        m_gradient += -(2/n) * x * (y- (m_now * x + b_now))
        b_gradient += -(2/n) * (y- (m_now * x + b_now))

    m = m_now - m_gradient* L 
    b = b_now - b_gradient* L 
    return m , b
    
```

```python
m =0 
b=0 
L = 0.0001
epochs = 50000 

for i in range(epochs):
    m,b = gradient_descent( m,b,data,L) 
    if i % 1000 == 0:
        current_loss = loss_function(m, b, data)
        print(f"Epoch {i}: m={m:.4f}, b={b:.4f}, Loss={current_loss:.4f}")

print(m,b) 
```

```python
x_min = data.StudyTime.min()
x_max = data.StudyTime.max()
plt.scatter(data.StudyTime, data.Score)
plt.plot([x_min, x_max], [m * x_min + b, m * x_max + b], color="red")
plt.show()
    
```
